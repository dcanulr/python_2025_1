### 3. Views and copys
- Views: Es una referencia a los mismos datos en memoria. Los cambios en view afectan a los datos originales
    * Regular index y slicing
- Copy: Es una copia de los datos. Los cambios en la copia no afectan a los datos originales
    * Fancy index
    * copy()

#### 3.1 Regular index
Usar índices enteros o slices

```python
arr = np.array([10, 20, 30, 40, 50])
arr[2]
```

```python
arr[1:5:2]
```

#### 3.2 Fancy indexing
Utilizar arrays como índices, e incluso condiciones

```python
arr[[2,4]]
```
