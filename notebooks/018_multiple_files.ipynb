## Importar dos archivos

```python
def read_ruoa(path):
    df = pd.read_csv(path, skiprows=[0,1,2,3,4,5,7], usecols=[0,1,2,3,8], 
            encoding= 'unicode_escape', index_col=0, parse_dates=True)
    return df
```

```python
path = "../data/2022-ene-tmix_hora_L1.csv"
df_ene = read_ruoa(path)
df_ene
```

```python
path = "../data/2022-feb-tmix_hora_L1.csv"
df_feb = read_ruoa(path)
df_feb
```

```python
path = "../data/2022-mar-tmix_hora_L1.csv"
df_mar = read_ruoa(path)
df_mar
```

## Resumen para importar múltiples archivos
1. Observar y analizar el archivo original (te le quedas viendo)
1. Importar un archivo de acuerdo a las necesidades
1. Crear una función para importar el mismo tipo archivos/archivos de la misma fuente
1. Utilizar glob para tener una lista con las rutas de todos los archivos
1. Iterar sobre la lista anterior y leer los archivos con la función del paso 3
1. Concatenar y ordenar el axis

```python
def read_ruoa(path):
    df = pd.read_csv(path, skiprows=[0,1,2,3,4,5,7], usecols=[0,1,2,3,8], 
            encoding= 'unicode_escape', index_col=0, parse_dates=True)
    return df
```

```python
paths = glob("../data/2022*")
df = [read_ruoa(file) for file in paths]
```

```python
len(df)
```

```python
df = pd.concat(df)
```

```python
df.sort_index(inplace=True)
df
```

```python
df.to_csv("../data/temixco_ruoa_2022.csv")
```
